```python
#creating a function to generate random numbers
from random import SystemRandom, random
def random_generate(n,secure):
  lst = []
  if secure ==True:
    random_number = SystemRandom().random()
  else:
    random_number = random()
  for i in range(n):
    lst.append(random_number)
  return lst
print("secure: ",random_generate(10,True))
print("unsecure: ",random_generate(10,False))

```

check time takes for both functions

```python
%%timeit
random_generate(100,True)#SystomRandom function : slowest but highly secure
```

```python
%%timeit
random_generate(100,False)#random function : average
```

Password generator using numpy choice function

```python
#password generator using random module
from random import choice
def password_generator(length):
  charectors = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#"
  password = ""
  for i in range(length):
    password += choice(charectors)
  return password
print(password_generator(10))
```

**True Random Number** : The truth is that most random numbers used in computer
programs are pseudo-random. The numbers are generated in a predictable way, because the algorithm is
deterministic. Pseudo-random numbers are good enough for many purposes, but it may not be "true" random
rolling dice or lottery drawings.

**Creating a Rock Papper Scissor game...**

```python
from random import choice

choices = ["Rock", "Papper", "Scissor"]
while(True):
  bot = choice(choices)
  user = int(input("0. Exit \n1. Rock\n2. Papper\n3. Scissor\nEnter: "))
  if user == 0:
    break
  elif user == 1:
    user = "Rock"
  elif user == 2:
    user = "Papper"
  elif user == 3:
    user = "Scissor"
  else:
    print("Invalid input")
    continue

  print(f"bot choice: {bot}")
  print(f"user choice: {user}")

  if user == bot:
    print("Draw..")
  elif user == "Rock" and bot == "Scissor" or user == "Scissor" and bot == "Papper" or  user == "Papper" and bot == "Rock":
    print("User win")
  else:
    print("Bot win")


```
